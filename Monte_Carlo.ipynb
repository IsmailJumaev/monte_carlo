```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = yf.download('TSLA')
```

```python
returns = np.log(1+df['Adj Close'].pct_change())
returns
```

```python
mu , signa = returns.mean(), returns.std()
```

```python
sim_rets = np.random.normal(mu, signa,252)
```

```python
initial = df['Adj Close'].iloc[-1]
initial
```

```python
sim_prices = initial * (1+sim_rets).cumprod()
```

```python
plt.plot(sim_prices)
```

```python
for i in range(100):
    sim_rets = np.random.normal(mu, signa,252)
    sim_prices = initial * (1+sim_rets).cumprod()
    plt.axhline(y=initial, color='r', linestyle='--')
    plt.plot(sim_prices)
```

```python
# prompt: I need to calculate ROI of Apple inc

import yfinance as yf

ticker = "AAPL"
start_date = "2022-01-01"  # Replace with your desired start date
end_date = "2022-12-31"    # Replace with your desired end date

# Download historical data
data = yf.download(ticker, start=start_date, end=end_date)

# Calculate initial and final prices
initial_price = data['Adj Close'].iloc[0]
final_price = data['Adj Close'].iloc[-1]

# Calculate ROI
roi = (final_price - initial_price) / initial_price

print(f"ROI for {ticker} from {start_date} to {end_date}: {roi:.2%}")
```
